```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException

import time
import numpy as np
import logging

import string
import random
```

```python
# Set up Selenium driver and navigate to online store
# driver = webdriver.Chrome()
driver = webdriver.Firefox()

# Define mean and standard deviation of order volume and product popularity based on historical data
mean_order_volume = 1.7
stddev_order_volume = 1
mean_product_popularity = 1000
stddev_product_popularity = 200

# Set up random seed for reproducibility
# np.random.seed(123)

# Generate popularity scores for each product
product_popularities = np.round(np.random.normal(
    loc=mean_product_popularity, scale=stddev_product_popularity, size=100)).astype(int)

# Generate list of product URLs based on their IDs and popularity scores
product_urls = ["http://localhost/prestashop/index.php?id_product=" +
                str(i) + "&controller=product" for i in range(1, 101)]
product_probs = product_popularities / np.sum(product_popularities)

# Loop through each order volume and modify product quantities
n_orders = 1000000
order_url = "http://localhost/prestashop/index.php?controller=order"

for i in range(n_orders):

    # Select random products and quantities to add to cart
    order_volume = int(np.random.normal(
        loc=mean_order_volume, scale=stddev_order_volume))
    if order_volume < 1:
        order_volume = 1
    product_selections = np.random.choice(
        product_urls, size=order_volume, replace=True, p=product_probs)
    np.random
    try:
        for product_url in product_selections:
            driver.get(product_url)  # navigate to product page
            # Modify quantity of product
            quantity_field = driver.find_element(By.NAME, "qty")
            # generate new quantity value
            new_quantity = int(np.random.uniform(1, 2))
            # quantity_field.send_keys(str(new_quantity))  # input new quantity value
            driver.execute_script(
                f"document.getElementById('quantity_wanted').value ={str(new_quantity)}")

            time.sleep(1)

            # adding order to cart
            button = driver.find_element(
                By.CLASS_NAME, "btn.btn-primary.add-to-cart")
            button.click()
            time.sleep(1)

        # go to order page
        time.sleep(1)
        driver.get(order_url)
        # personal information form

        # name
        name = driver.find_element(By.ID, "field-firstname")
        name.clear()
        name.send_keys(''.join(random.choices(string.ascii_uppercase, k=9)))

        # surname
        surname = driver.find_element(By.ID, "field-lastname")
        surname.clear()
        surname.send_keys(''.join(random.choices(string.ascii_uppercase, k=9)))

        # mail
        mail = driver.find_element(By.ID, "field-email")
        mail.clear()
        mail.send_keys(f"{np.random.randint(1000000)}@mail.pl")

        # check first agree
        check_box1 = driver.find_element(By.NAME, "customer_privacy")
        check_box1.click()

        # check second agree
        check_box2 = driver.find_element(By.NAME, "psgdpr")
        check_box2.click()

        # go to next page
        data_confirm = driver.find_element(By.NAME, "continue")
        data_confirm.click()

        # formularz adres
        try:
            elem = driver.find_element(
                By.CLASS_NAME, "edit-address.text-muted")
            if elem.is_displayed():
                elem.click()  # this will click the element if it is there
                time.sleep(1)
                street = driver.find_element(By.ID, "field-address1")
                street.clear()
                street.send_keys(np.random.randint(1000000))

                city = driver.find_element(By.ID, "field-city")
                city.clear()
                city.send_keys(np.random.randint(1000000))

                post_code = driver.find_element(By.ID, "field-postcode")
                post_code.clear()
                post_code.send_keys(
                    f"50-{np.random.randint(low= 100,high= 999)}")

                adres_confirm = driver.find_element(
                    By.NAME, "confirm-addresses")
                adres_confirm.click()

        except NoSuchElementException:
            street = driver.find_element(By.ID, "field-address1")
            street.clear()
            street.send_keys(np.random.randint(1000000))

            city = driver.find_element(By.ID, "field-city")
            city.clear()
            city.send_keys(np.random.randint(1000000))

            post_code = driver.find_element(By.ID, "field-postcode")
            post_code.clear()
            post_code.send_keys(f"50-{np.random.randint(low= 100,high= 999)}")

            adres_confirm = driver.find_element(By.NAME, "confirm-addresses")
            adres_confirm.click()

        # driver.find_element(By.CLASS_NAME, "edit-address.text-muted")

        # street = driver.find_element(By.ID, "field-address1")
        # street.clear()
        # street.send_keys(np.random.randint(1000000))

        # city = driver.find_element(By.ID, "field-city")
        # city.clear()
        # city.send_keys(np.random.randint(1000000))

        # post_code = driver.find_element(By.ID, "field-postcode")
        # post_code.clear()
        # post_code.send_keys(f"50-{np.random.randint(low= 100,high= 999)}")

        # adres_confirm = driver.find_element(By.NAME, "confirm-addresses")
        # adres_confirm.click()

        # formularz dostawa
        delivery_confirm = driver.find_element(
            By.NAME, "confirmDeliveryOption")
        delivery_confirm.click()

        # formularz płatność
        payment_option = driver.find_element(By.ID, "payment-option-2")
        payment_option.click()
        last_agree = driver.find_element(
            By.ID, "conditions_to_approve[terms-and-conditions]")
        last_agree.click()
        time.sleep(1)
        # payment_confirm = driver.find_element(By.NAME, "confirm-addresses")
        # payment_confirm.click()
        make_order = driver.find_element(
            By.CLASS_NAME, "btn.btn-primary.center-block")
        make_order.click()

        # Wait for a few seconds before proceeding to the next order
        time.sleep(1)

    # except Exception as e: print(e)
    except BaseException:
        logging.exception("An exception was thrown!")


# Close Selenium driver
driver.quit()
```
